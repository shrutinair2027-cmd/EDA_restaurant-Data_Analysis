```python
restaurant_name=[]
restaurant_location=[]
res_details=[]
res_offers=[]

for i in range(1,5):
    url=f"https://www.eazydiner.com/restaurants?date=2026-03-02&meal_period=dinner&location=ahmedabad&page={i}"
    headers={"user-agent":
             "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"}

    res=requests.get(url,headers=headers)
    soup=BeautifulSoup(res.text)
    containers=soup.find_all("div",attrs={"class":"flex align-v-start flex-between padding-t-15 padding-b-15 margin-t-12 relative full-width"})
    for c in containers:
        r_name=c.find("div",attrs={"class":"relative bold ellipsis listing_res_name__uVIN8"})
        if r_name is None:
            restaurant_name.append(np.nan)
        else:
            restaurant_name.append(r_name.text)

        r_location=c.find("div",attrs={"class":"flex align-v-center gap_5 padding relative"})
        if r_location is None:
            restaurant_location.append(np.nan)
        else:
            restaurant_location.append(r_location.text)

        r_details=c.find("div",attrs={"class":"flex align-v-center gap_5 relative"})
        if r_details is None:
            res_details.append(np.nan)
        else:
            res_details.append(r_details.text)

        offers=soup.find("div",attrs={"class":"relative font-16 bold white nowrap ellipsis"})
        if offers is None:
            res_offers.append(np.nan)
        else:
            res_offers.append(offers.text)    
```
